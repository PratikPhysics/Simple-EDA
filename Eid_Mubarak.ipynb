```python
# pratikphysics1991@gmail.com
# https://github.com/PratikPhysics/

import turtle
import random

# Function to draw stars
def draw_star(x, y):
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.color("white")
    pen.begin_fill()
    for _ in range(5):
        pen.forward(10)
        pen.right(144)
    pen.end_fill()

# Function to draw a mosque silhouette
def draw_mosque(x, y):
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.color("black")
    pen.begin_fill()
    pen.goto(x, y + 50)
    pen.goto(x + 100, y + 100)
    pen.goto(x + 200, y + 50)
    pen.goto(x + 200, y)
    pen.goto(x, y)
    pen.end_fill()


# Function to draw a circle
def draw_circle(color, radius, x, y):
    pen.penup()
    pen.goto(x, y)
    pen.pendown()
    pen.color(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()

# Create a Turtle screen
screen = turtle.Screen()
screen.setup(width=600, height=400)
screen.bgcolor("gray")

# Create a Turtle object
pen = turtle.Turtle()
pen.speed(1)  # Slowing down the turtle
pen.penup()
pen.color("green")
pen.goto(-100, 0)

# Write "Eid Mubarak"
pen.write("Eid Mubarak", font=("Arial", 36, "bold"))



# Draw the white circle
draw_circle("white", 50, 10,60)

# Draw the gray moon (forming a crescent moon shape)
draw_circle("gray", 50, 20,68)



# Draw mosque silhouette in the background
draw_mosque(-250, -150)


# Draw stars randomly on the screen
for _ in range(10):
    x = random.randint(-300, 300)
    y = random.randint(-200, 200)
    draw_star(x, y)


# Hide the turtle
pen.hideturtle()

# Keep the window open
screen.mainloop()
```

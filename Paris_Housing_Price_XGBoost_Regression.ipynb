## Paris Housing Price Prediction
[https://www.kaggle.com/datasets/mssmartypants/paris-housing-price-prediction/data]


```python
#Importing libraries
import xgboost as xgb
import seaborn as sns

import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Loading the dataset
df = pd.read_csv('ParisHousing.csv')
df
```

```python
# droping the cityCode column
df = df.drop(columns='cityCode')
```

```python
# Creating an x and y
x = df.iloc[:, :-1].values
y = df.price.values
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Converting the data into DMatrix format, which is optimized for XGBoost
dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)

# Defineing parameters for the XGBoost model
params = {
    'objective': 'reg:squarederror',  # Specify the objective as regression
    'max_depth': 3,                    # Maximum depth of each tree
    'learning_rate': 0.1              # Learning rate (step size) for each boosting step

}

# Training the XGBoost model
model = xgb.train(params, dtrain)

# Makeing predictions on the test set
predictions = model.predict(dtest)

# Calculating mean squared error (MSE) to evaluate the model's performance
mse = mean_squared_error(y_test, predictions)
print("Mean Squared Error:", mse)
```
